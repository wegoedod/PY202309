```python
import random

def get_number(): #1~45까지의 값을 무작위로 반환하는 함수
    return random.randrange(1, 46) 

def set_lotto_numbers():
    lotto = [] #lotto 리스트 생성
    while True:
        num = get_number()
    
        if lotto.count(num) == 0: #num이 lotto안에 없을 경우
            lotto.append(num)

        if len(lotto) >= 6: #lotto의 길이가 6 보다 크거나 같을 경우 while문 탈출
            break

    return lotto

def match_numbers(user_inputs, lotto_numbers): #같은 값이 있을 때 마다 1을 더하는 for문
    match = 0
    for i in user_inputs:
        if i in lotto_numbers:
            match +=1
    return match
    
def check_prize(same_numbers): #same_numbers에 따라 등수 출력
    if same_numbers == 6:
        print("First prize")
    elif same_numbers == 5:
        print("Second prize")
    elif same_numbers == 4:
        print("Third prize")
    elif same_numbers == 3:
        print("Fourth prize")
    else:
        print("no prize :(")
    
    
num = 0
print("** start Lotto **")
print("1) Set the lotto numbers")
lotto_numbers = set_lotto_numbers()

print("2) Set the user inputs ") #유저가 입력한 대로 user_inputs에 저장(범위 1~46 중복 허용 X)
user_inputs = []
while 1:
    num  = int(input("Enter %d-number: "% (len(user_inputs)+1)))
    if num < 1 or num > 46:
        print("No valid number")
        continue
    if user_inputs.count(num) == 0:
        user_inputs.append(num)
    else:
        print("Same number")
    if len(user_inputs) >= 6:
        break
        
print("3) Match the numbers") 
same_numbers = match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize")
check_prize(same_numbers)

print("5) Sort the lists") #두 리스트 정렬
lotto_numbers.sort()
user_inputs.sort()

print("=== Summary ===")
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs)
```
