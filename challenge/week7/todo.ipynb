```python
def add_task(task_list, task): //task_list에 task를 추가함
    task_list.append(task+"\n")
def display_tasks(task_list)://task_list를 출력함
    for i in task_list:
        print(i,end="")
    print()
def save_tasks_to_file(task_list, filename)://task_list를 파일에 씀
    fp = open(filename, "w", encoding="utf8")
    fp.writelines("".join(task_list))
    fp.close()
def load_tasks_from_file(filename)://파일을 불러옴
    task_list = []
    fp = open(filename, "r", encoding="utf8")
    task_list = fp.readlines()
    fp.close()
    return task_list
file_name = "task.txt"
task_list = load_tasks_from_file(file_name)

while True: //사용자의 입력을 받아서 수행할 일을 선택
    print("\n할일 관리 프로그램")
    print("1. 할일 추가")
    print("2. 할일 목록 표시")
    print("3. 종료")
    choice = input("선택: ")
    print()
    if choice == '1':
        task = input("할일을 입력하세요: ") //task 입력 받음
        add_task(task_list, task) //add_tasks함수 호출
        print("할일이 추가되었습니다.")
    elif choice == '2':
        display_tasks(task_list)//display_tasks함수 호출
    elif choice == '3':
        save_tasks_to_file(task_list, file_name) //save_tasks_to_file함수 호출
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```
